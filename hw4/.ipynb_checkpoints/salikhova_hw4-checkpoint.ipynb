# Теоретическое введение


Основная задача кластерного анализа — разбиение исходного набора объектов на группы (кластеры) таким образом, чтобы объекты в группе были похожи друг на друга, а объекты из разных групп - отличались. 

В этой работе мы будем реализовывать агломеративную кластеризацию. 

## Агломеративная кластеризация

Иерархические алгоритмы кластеризации строят на выборке систему вложенных разбиений.  Наиболее часто испоьзуемые иерархические алгоритмы - агломеративные алгоритмы, которые строят ирархию по восходящей, то есть от момента, когда все элементы являются отдельными кластерами, до того момента, пока вся выборка не станет одним кластером.

Сам алгоритм описать можно примерно так:

* начинаем с ситуации, когда каждый объект - отдельный кластер
* на каждом шаге совмещаем два наиболее близких кластера
* останавливаемся, когда получаем требуемое количество или единственный кластер


В данной работе Вам предлагается реализовать три варианта расстояний между кластерами, по которым Вы определяете пару для слияния: 
* Single linkage
$$ d_{min}(C_i, C_j) = \min_{\mathbf{x} \in C_i, \mathbf{x}' \in C_j} \|\mathbf{x} -\mathbf{x}' \| $$

* Complete linkage
$$ d_{max}(C_i, C_j) = \max_{\mathbf{x} \in C_i, \mathbf{x}' \in C_j} \|\mathbf{x} -\mathbf{x}' \| $$

* Average linkage
$$ d_{avg}(C_i, C_j) = \frac{1}{n_i n_j}\sum_{\mathbf{x} \in C_i}\sum_{\mathbf{x}' \in C_j} \|\mathbf{x} -\mathbf{x}' \| $$

Как считать расстояние $\|\mathbf{x} -\mathbf{x}' \|$ между объектами Вы должны выбрать сами.  Вы должны четко понимать, что выбор расстояния между объектами - самое важное в такого рода алгоритмах. Передавайте название расстояния в конструктор класса, а потом, в зависимости от названия, применяете ту или иную функцию для подсчета расстояния. 



Для того, чтобы на втором этапе нам не пересчитывать заново расстояния между всеми парами кластеров, можно использовать Формулы Ланса-Вильямса, которые позволяют посчитать расстояние между кластерами после объединения, используя только расстояния между кластерами до объедения.  Таким образом, достаточно будет посчитать расстояния только между всеми одноэлементными кластерами (первый этап), а затем пересчитывать расстояния при их объединении через 
формулы Ланса-Вильямса. 

Общая  формула выглядит так:
$$ d(C_i \cup C_j, C_k) = a_i \cdot d(C_i, C_k) + a_j \cdot d(C_j, C_k) + b \cdot d(C_i, C_j) + c \cdot |d(C_i, C_k) - d(C_j, C_k)|$$

Можно показать, что все относительно разумные кластерные расстояния можно описать данной формулой. 

Для интересущих нас linkage:

* single-linkage       $a_i = \frac{1}{2}, a_j =  \frac{1}{2}, b = 0, c =  - \frac{1}{2} $

* complete-linkage     $a_i = \frac{1}{2}, a_j =  \frac{1}{2}, b = 0, c =   \frac{1}{2} $

* average-linkage       $a_i = \frac{|C_i|}{|C_i \cup C_j|}, a_j =  \frac{|C_j|}{|C_i \cup C_j|}, b = 0, c = 0 $

Задача 1. Сходимость kmeans.

Докажите, что kmeans всегда сходится, и итераций не будет превышать $K^N$, где $K$ $-$ число кластеров, $N$ $-$ число объектов

Подсказка: Вам стоит рассмотреть функционал, который минимизирует алгоритм, и подумать, как он изменяется на E шаге и на M шаге.



**Решение:**


На Е шаге определяется принадлежность объектов класетрам - $r_{nk}$, на М шаге - пересчитываются положения центроидов - $\mu_{k}$ (при фиксированных $r_{nk}$) 

$$Q = \sum_{n=1}^N \sum_{k=1}^K r_{nk}\|x_n - \mu_k\|^2 \rightarrow min$$

Знаем, что минимум достигается на центроидах, которые соответствуют центрам масс кластеров. Алгоритм сходится, если сумма расстояний между элементами кластера и его центром масс убывает на каждом шаге алогритма. Пусть $\mu_0$ - предыдущий центр масс, $\mu_1$ - новый центр масс:

$$\sum_{i=1}^n\|x_i-\mu_0\|^2=\sum_{i=1}^n<x_i-\mu_0,x_i-\mu_0>=$$

$$=\sum_{i=1}^n<|x_i-\mu_1|+|\mu_1-\mu_0|,|x_i-\mu_1|+|\mu_1-\mu_0|>=$$

$$=\sum_{i=1}^n\|x_i-\mu_1\|^2+n\|\mu_1-\mu_0\|^2+2<|\mu_1-\mu_0|,\sum_{i=1}^n|x_i-\mu_1|>$$

$$\Rightarrow\sum_{i=1}^n\|x_i-\mu_1\|^2\lt\sum_{i=1}^n\|x_i-\mu_0\|^2$$

Число способов разбить множество из N элементов на K частей = $K^N$. Если на очередном шаге мы получили то же разбиение, что и на предыдущем(это случится, так как количество способов разбиения конечно), то это значит, что на М шаге мы получим то же значение центроиды и т.д. Т.е. алгоритм остановится и сделает максимум $K^N$ итераций.

Но при большом количестве данных центры будут перемещаться в малом радиусе, и элементы по краям кластеров будут прыгать то в один, то в другой кластер. Для этого предусмотрим в алгоритме максимальное число итераций.

Задача 2.  ММП для многомерного нормального распределения.

Функция правдоподобия в общем случае записывается как:
￼

Для многомерного нормального распределения плотность вероятности:
$$p(x_i| \theta) = \frac{1}{(2\pi)^{n/2}|\Sigma|^{1/2}}e^{-\frac{1}{2} (x_i - \mu)^T\Sigma^{-1}(x_i - \mu)}$$

Выведите ММП оценку для вектора $\mu$
￼

**Решение:**


Пусть c $=\frac{1}{(2\pi)^{n/2}|\Sigma|^{1/2}}$

$lnL=c\sum_{i=1}^n-\frac{1}{2}(x_i-\mu)^T\Sigma^{-1}(x_i-\mu)\Rightarrow\frac{\partial L}{\partial\mu}=\frac{c}{2}\sum_{i=1}^{n}(\Sigma^{-1}+(\Sigma^{-1})^{T})(x_i-\mu)=0\Rightarrow n\mu=\sum_{i=1}^n x_i$

$\Rightarrow \mu=\sum_{i=1}^n\frac{x_i}{n}$ 


Задача 3. Формулы Ланса-Вильямса.

Формулы Ланса-Вильямса крайне удобны для быстрого пересчта расстояний в агломеративной кластеризации. Давайте докажем  формулы для single и complete linkage. 

Докажите, что:

$d_{\min}(U \cup V, S) = \frac{1}{2} d_{\min}(U, S) + \frac{1}{2} d_{\min}(V, S)  - \frac{1}{2} | d_{\min}(U, S) - d_{\min}(V, S) |  $


$d_{\max}(U \cup V, S) = \frac{1}{2} d_{\max}(U, S) + \frac{1}{2} d_{\max}(V, S)  + \frac{1}{2} | d_{\max}(U, S) - d_{\max}(V, S) |  $




**Решение:**

Известно, что минимум и максимум двух чисел можно найти по формулам: 
$$min(a, b) = \frac{a + b - |a - b|}{2} $$
$$max(a, b) = \frac{a + b + |a - b|}{2} $$
Из этих соотношений и вытекает формулы для single(complete), так как мы ищем минимум(максимум) из расстояний $d(U, S)$ и $d(V, S)$, ведь $d(U, S)$ и $d(V, S)$ - есть минимальное(максимальное) расстояние между элементами множеств U и S, V и S.

# Реализация алгоритма кластеризации  (5 баллов)

Алгоритм кластеризации должен удовлетворять следующему интерфейсу. Конструктор принимает набор параметров, необходимых для работы алгоритма кластеризации. Метод `fit` подсчитывает параметры модели и возвращает `self`. Метод `predict` возвращает вектор с индексами кластеров для поданных в него объектов `x`. 

```python
from scipy.cluster import hierarchy
```

```python
from sklearn.base import BaseEstimator, ClusterMixin

class Clustering(BaseEstimator, ClusterMixin):
  
    def __init__(self, metric='euclidean',
                 linkage='complete', n_claster=2):
        self.metric = metric
        self.linkage = linkage
        self.k = n_claster
        return
        
    def fit_predict(self, X):
        dists2 = []
        n_iter = 0
        n_clasters = X.shape[0]
        dists1_1 = ss.distance.pdist(X, self.metric)
        dists1_2 = ss.distance.squareform(dists1_1)
        unsquareform = lambda a: a[np.nonzero(np.triu(a))]
        idx = list(combinations(range(X.shape[0]), 2))
        X_clasterization = [[i] for i in range(X.shape[0])]
        claster_idx = list(range(X.shape[0]))
        while n_clasters > self.k:  
            ind = np.argmin(dists1_1)
            ind1, ind2 = idx[ind]
            n_iter += 1
            idx = list(combinations(range(X.shape[0] - n_iter), 2))
            dists2 = dists1_2
            if self.linkage == 'single':
                a1 = 1 / 2
                a2 = 1 / 2
                b = 1 / 2
            elif self.linkage == 'complete':
                a1 = 1 / 2
                a2 = 1 / 2
                b = - 1 / 2
            else:
                c1 = len(X_clasterization[ind1])
                c2 = len(X_clasterization[ind2])
                a1 = c1 / (c1 + c2)
                a2 = c2 / (c1 + c2)
                b = 0
            dists2[:, ind1] = a1 * dists1_2[:, ind1] + a2 * dists1_2[:, ind2] +\
            b * abs(dists1_2[:, ind1] - dists1_2[:, ind2])
            dists2[ind1, :] = dists2[:, ind1]
            dists2 = np.delete(dists2, ind2, axis=1)
            dists2 = np.delete(dists2, ind2, axis=0)
            X_clasterization[ind1] += X_clasterization[ind2]
            X_clasterization = np.delete(X_clasterization, ind2)
            n_clasters -= 1
            if dists2 == dists1_2:
                break
            else:
                dists1_1 = unsquareform(dists2)
                dists1_2 = dists2
        if len(X_clasterization) != self.k:
            raise RuntimeError('number of clasters!!!')
        predict = zeros(X.shape[0]) 
        for i in range(self.k):
            predict[X_clasterization[i]] = i
        return predict.astype(int)
    
    def plot_dendrogram(self, X):   
        dists2 = []
        n_iter = 0
        n_clasters = X.shape[0]
        dists1_1 = ss.distance.pdist(X, self.metric)
        dists1_2 = ss.distance.squareform(dists1_1)
        unsquareform = lambda a: a[np.nonzero(np.triu(a))]
        idx = list(combinations(range(X.shape[0]), 2))
        X_clasterization = [[i] for i in range(X.shape[0])]
        linkage_mtx = []
        claster_idx = list(range(X.shape[0]))
        while n_clasters > self.k-1:  
            ind = np.argmin(dists1_1)
            ind1, ind2 = idx[ind]
            linkage_mtx.append([claster_idx[ind1],
                                claster_idx[ind2], dists1_1[ind],
                                len(X_clasterization[ind1]) +\
                                len(X_clasterization[ind2])])
            claster_idx[ind1] = X.shape[0] + n_iter
            claster_idx = np.delete(claster_idx, ind2)
            n_iter += 1
            idx = list(combinations(range(X.shape[0] - n_iter), 2))
            dists2 = dists1_2
            if self.linkage == 'single':
                a1 = 1 / 2
                a2 = 1 / 2
                b = 1 / 2
            elif self.linkage == 'complete':
                a1 = 1 / 2
                a2 = 1 / 2
                b = - 1 / 2
            else:
                c1 = len(X_clasterization[ind1])
                c2 = len(X_clasterization[ind2])
                a1 = c1 / (c1 + c2)
                a2 = c2 / (c1 + c2)
                b = 0
            dists2[:, ind1] = a1 * dists1_2[:, ind1] + a2 * dists1_2[:, ind2] +\
            b * abs(dists1_2[:, ind1] - dists1_2[:, ind2])
            dists2[ind1, :] = dists2[:, ind1]
            dists2 = np.delete(dists2, ind2, axis=1)
            dists2 = np.delete(dists2, ind2, axis=0)
            X_clasterization[ind1] += X_clasterization[ind2]
            X_clasterization = np.delete(X_clasterization, ind2)
            n_clasters -= 1
            dists1_1 = unsquareform(dists2)
            dists1_2 = dists2
        if len(X_clasterization) != self.k-1:
            raise RuntimeError('number of clasters!!!')
        plt.figure(figsize = (20, 5))
        hierarchy.dendrogram(linkage_mtx)
        plt.show()
```

Сначала проверим реализованный алгоритм кластеризации на классическом наборе данных [Iris](http://www.wikiwand.com/en/Iris_flower_data_set). Загрузим данные (они включены в библиотеку sklearn) и посмотрим на то, как они выглядят в двух проекциях (для простоты используем 2 класса из 3).
